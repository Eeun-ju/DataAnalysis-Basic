# SOM(Self-Organizing Map)

     SOM이란?
     차원이 축소화된 데이터셋을 topology 개념으로 효과적인 표현을 만들 수 있는 기법이다. 비지도학습 기법 중 하나로 주로 군집화에 많이 사용된다.
     
     SOM 아키텍처
     n차원 입력벡터의 요소들을 2차원 격자에 표현한다고 하자. 저차원 격자 하나에는 여러 개의 입력벡터들이 속할 수 있는데, 동일한 격자에 속하는 벡터들은 위치적인 유사도를 가진다. 전체적인 흐름은 임의의 n차원 입력벡터가 들어왔을 때 가장 가까운 격자벡터(Winning node)를 찾아 군집화를 이루는 것이다. 같은 저차원 격자에 할당된 벡터들은 Winning node와의 거리가 제각기 다를 수 있다. 
     
     SOM 학습 방법
     원데이터 공간에서 격자벡터를 랜덤으로 초기화 하고 입력벡터를 하나씩 추가하면서 격자벡터 위치를 업데이트 한다. 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn.cluster as cluster
import matplotlib.pyplot as plt
```

```python
x = np.random.randn(10000)
y = np.random.randn(10000)

df = pd.DataFrame({"X":x,"Y":y})
df['dis'] = df['X']**2 + df['Y']**2
df['cla'] = df['dis'].apply(lambda x:1 if 0.5<x<1.0 else 2 if 3<x<4 else 0)
df['cla'].value_counts() # Series에서 변수 count 가능
df = df[df['cla']!=0]
```

```python
sns.scatterplot(df['X'],df['Y'],hue = df['cla'])
```

```python
from somlearn import SOM
clusterer = SOM(n_columns=1,n_rows=2,gridtype='hexagonal',random_state = 70)
ypred = clusterer.fit_predict(df)
df['pred'] = ypred
sns.scatterplot(df['X'],df['Y'],hue=df['pred'])
```
