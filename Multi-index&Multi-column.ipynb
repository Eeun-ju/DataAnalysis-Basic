```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
#데이터 불러오기
iris = load_iris()
#print(iris.DESCR)
```

```python
data = pd.DataFrame(data = iris.data, columns = iris.feature_names)
data['target'] = iris.target
```

```python
data.head()
```

### 0,1,2 label을 문자열로 매핑하는 방법

```python
data['target'] = data['target'].map({0:"setosa",1:"versicolor",2:"virginica"})
data.head()
```

```python
# index이름 세팅하기 여러 인덱스로 설정 가능 ['idex1','index2']
df1 = data.set_index('target')
df1.head()
```

```python
df2 = pd.DataFrame(np.arange(16).reshape(4,4),
                         index = [[1,1,2,2],['s1','s2','s1','s2']],
                         columns = [['apple','apple','banana','banana'], ['price','qty','price','qty']],
                         )
df2
```

```python
df2.xs('apple',axis=1) #같은 표현 df2['apple']
```

```python
df2.xs('apple',level = 0, axis = 1) #axis = 0,1(row,column) level = 0,1(상위, 하위)
```

```python
df2.xs('apple',level = 0, axis = 1).xs(2) # index 접근은 xs이용
```

## 계층 순서 변경

```python
df1 = pd.DataFrame(np.arange(16).reshape(4,4), index=[['a','a','b','b'], [1,2,1,2]],
columns = [['col1','col1','col2','col2'], ['c1','c2','c1','c2']])
df1
```

```python
df1.swaplevel(0,1) #계층 순서 변경 index
```

```python
df1.swaplevel(0,1,axis = 1) #column 계층 순서 변경
```
